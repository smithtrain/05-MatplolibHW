```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
df_city = pd.read_csv(city_data_to_load)
df_ride= pd.read_csv(ride_data_to_load)

#print(df_city.head())
#df_ride.head()

#print(len(df_city['city']))


#print(len(df_ride['date']))

# Combine the data into a single dataset

df_data = df_ride.merge(right=df_city, how='inner', on='city')

# Display the data table for preview
df_data.head()
print(len(df_data['city']))
#print(len(df_data['date']))

#df_data.to_excel('Data.xlsx')
```

```python
# Obtain the x and y coordinates for each of the three city types
x_data_gpd = df_data.groupby('city')['ride_id'].count()
y_data_gpd = df_data.groupby('city')['fare'].mean()
fare_sum_gpd = df_data.groupby('city')['fare'].sum()
bubble_size_sqd_gpd = (df_data.groupby('city')['driver_count'].sum())/3.14592
bubble_size_gpd = bubble_size_sqd_gpd**(0.50)

df_x_data_gpd = pd.DataFrame(x_data_gpd)
df_y_data_gpd = pd.DataFrame(y_data_gpd)
df_fare_sum_gpd = pd.DataFrame(fare_sum_gpd)
df_bubble_size_sqd_gpd = pd.DataFrame(bubble_size_sqd_gpd)
df_bubble_size_gpd = pd.DataFrame(bubble_size_gpd)

df_data_grpd = df_x_data_gpd.merge(right=df_y_data_gpd, how='inner', on='city')
df_data_grpd = df_data_grpd.merge(right=df_fare_sum_gpd, how='inner', on='city')
df_data_grpd = df_data_grpd.merge(right=df_bubble_size_sqd_gpd, how='inner', on='city')
df_data_grpd = df_data_grpd.merge(right=df_bubble_size_gpd, how='inner', on='city')
df_data_grpd = df_data_grpd.merge(right=df_city, how='inner', on='city')
df_data_grpd = df_data_grpd.drop('driver_count',axis='columns')

df_data_grpd = df_data_grpd.rename(columns={'ride_id':'Ride_Cnt','fare_x':'Avg_Fare','fare_y':'Sum_Fare','driver_count_x':'Driver_Cnt_Sqd','driver_count_y':'Driver_Cnt'})
df_data_grpd.head()
```

## Bubble Plot of Ride Sharing Data

```python

city_types = ['Urban','Suburban', 'Rural']

print(city_types)

for city_type in city_types:    
    
    if city_type =='Urban': 
        
        x_values = df_data_grpd.loc[(df_data_grpd['type']=='Urban')]['Ride_Cnt']        
        y_values  = df_data_grpd.loc[(df_data_grpd['type']=='Urban')]['Avg_Fare']    
        color_val = "#F08080" #(0.240, 0.128, 0.128)  #LightCoral #F08080	RGB(240,128,128)
        size_values = df_data_grpd.loc[(df_data_grpd['type']=='Urban')]['Driver_Cnt_Sqd']
        
    elif city_type =='Suburban':
        
        x_values = df_data_grpd.loc[(df_data_grpd['type']=='Suburban')]['Ride_Cnt']
        y_values  = df_data_grpd.loc[(df_data_grpd['type']=='Suburban')]['Avg_Fare']     
        color_val = "#87CEFA" #(0.135, 0.206, 0.250) #'LightSkyBlue' lightskyblue	#87CEFA	RGB(135,206,250)
        size_values = df_data_grpd.loc[(df_data_grpd['type']=='Suburban')]['Driver_Cnt_Sqd']       
        
    else:
        df_data_Rur_grpd_ = df_data_grpd.loc[(df_data_grpd['type']=='Rural')]
        x_values = df_data_grpd.loc[(df_data_grpd['type']=='Rural')]['Ride_Cnt']
        y_values  = df_data_grpd.loc[(df_data_grpd['type']=='Rural')]['Avg_Fare']     
        color_val = "#FFD700" #(.255, .215, 0.000) #'Gold' #gold1	#FFD700	RGB(255,215,0)
        size_values = df_data_grpd.loc[(df_data_grpd['type']=='Rural')]['Driver_Cnt_Sqd']         
 
    plt.scatter(x_values, y_values, marker="o", c=color_val, edgecolors="black", s=size_values, alpha=0.60, label=city_type)
    



# Incorporate the other graph properties
plt.title("Pyber Ride Sharing (2016)")
plt.ylabel("Average Fare per ride ($)")
plt.xlabel("Total Number of Riders (per City)")
plt.grid()
plt.legend(loc='upper right', title="City Types",markerscale=.75)
plt.text(42, 30, s="Note: Bubble size correlates with driver count per city", fontsize=10)
plt.savefig("PyberRideSharing.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_grpd = df_data_grpd.groupby('type')['Sum_Fare'].sum()
df_fares_grpd = pd.DataFrame(fares_grpd)
df_fares_grpd_srt = df_fares_grpd.sort_values('Sum_Fare')
city_types = ['Rural','Suburban','Urban' ]

#print(df_fares_grpd_srt)




colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,0.1)

# Build Pie Chart
# Tell matplotlib to create a pie chart based upon the above data
plt.pie(df_fares_grpd_srt['Sum_Fare'], explode=explode, labels=city_types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
# Create axes which are equal so we have a perfect circle
plt.axis("equal")
plt.title("% of Total Fares by City Type")
# Save an image of our chart and print the final product to the screen
plt.savefig("Pyber_TotFaresbyCityType.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Type Percents
Rides_grpd = df_data_grpd.groupby('type')['Ride_Cnt'].sum()
df_rides_grpd = pd.DataFrame(Rides_grpd)
df_rides_grpd_srt = df_rides_grpd.sort_values('Ride_Cnt')
city_types = ['Rural','Suburban','Urban' ]

#print(df_fares_grpd_srt)




colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,0.1)

# Build Pie Chart
# Tell matplotlib to create a pie chart based upon the above data
plt.pie(df_rides_grpd_srt['Ride_Cnt'], explode=explode, labels=city_types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
# Create axes which are equal so we have a perfect circle
plt.axis("equal")
plt.title("% of Total Rides by City Type")
# Save an image of our chart and print the final product to the screen
plt.savefig("Pyber_TotRidesbyCityType.png")
```

```python
# Show Figure
plt.show()
```
